## –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ train.csv –∏ features.csv –Ω–∞ –æ—Å–Ω–æ–≤–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤

–ß—Ç–æ–±—ã —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏–∑ —Ñ–∞–π–ª–∞ `features.ipynb` –±—ã–ª–∏ –æ—Å–æ–∑–Ω–∞–Ω–Ω—ã–µ, –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –∏–∑ `features.csv` –±—ã–ª–∏ —Å–¥–µ–ª–∞–Ω—ã –¥–∏–∞–ø–∞–∑–æ–Ω—ã —Å —à–∞–≥–æ–º **5**, —á—Ç–æ–±—ã –º–æ–∂–Ω–æ –±—ã–ª–æ –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π —à–∏—Ä–æ—Ç—ã `(lat)` –∏ –¥–æ–ª–≥–æ—Ç—ã `(lon)` –ø—Ä–∏—Å–≤–æ–∏—Ç—å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

```python
# –®–∞–≥ —Ä–∞–∑–±–∏–µ–Ω–∏—è –Ω–∞ –¥–∏–∞–ø–∞–∑–æ–Ω—ã
step = 5
features['lat_bin'] = (features['lat'] // step) * step
features['lon_bin'] = (features['lon'] // step) * step

grouped_features_df = features.groupby(['lat_bin', 'lon_bin']).mean().reset_index()

train_data['lat_bin'] = (train_data['lat'] // step) * step
train_data['lon_bin'] = (train_data['lon'] // step) * step

merged_df = pd.merge(train_data, grouped_features_df, on=['lat_bin', 'lon_bin'], how='left')
merged_df.drop(['lat_bin', 'lon_bin'], axis=1, inplace=True)
```

## –ü–µ—Ä–µ—Å–±–æ—Ä–∫–∞ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞

```python
merged_df = merged_df[[col for col in merged_df.columns if col != 'score'] + ['score']]
merged_df = merged_df.drop(['lat_y', 'lon_y'], axis=1)
merged_df['id'] = range(0, 2000)
merged_df = merged_df.set_index('id')

merged_df.head()
```

# üë®üèª‚Äçüè´ –û–±—É—á–µ–Ω–∏–µ –±–∞–∑–æ–≤—ã—Ö –º–æ–¥–µ–ª–µ–π

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from catboost import CatBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor

X = merged_df.drop('score', axis=1)
y = merged_df['score']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=0)
```

–°–æ–∑–¥–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –∏ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –º–æ–¥–µ–ª–µ–π.

```python
def train_pipe_test(models, X_tr, X_te,
                    y_tr, y_te):
    set_models = {}
    for model in models:
        temp_pipe = Pipeline([('scaler', StandardScaler()),
                              (f"{type(model).__name__}", model)])
        temp_pipe.fit(X_tr, y_tr)

        set_models[type(model).__name__] = np.mean(np.abs(temp_pipe.predict(X_te) - y_te))

    return set_models

def train_pipe_train(models, X_tr, X_te,
                     y_tr, y_te):
    set_models = {}
    for model in models:
        temp_pipe = Pipeline([('scaler', StandardScaler()),
                              (f"{type(model).__name__}", model)])
        temp_pipe.fit(X_tr, y_tr)

        set_models[type(model).__name__] = np.mean(np.abs(temp_pipe.predict(X_tr) - y_tr))

    return set_models
```

```python
%%capture
models_list = [LinearRegression(), Lasso(), CatBoostRegressor(),
               KNeighborsRegressor(), DecisionTreeRegressor(), Ridge(), RandomForestRegressor()]

results_train = train_pipe_train(models_list, X_train, X_test, y_train, y_test)
results_test = train_pipe_test(models_list, X_train, X_test, y_train, y_test)
```

### –ó–Ω–∞—á–µ–Ω–∏–µ MAE –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –∏ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö

```python
print('Train MAE')
results_train
```

```python
print('Test MAE')
results_test
```

```python
def plot_dicts(dict1, dict2):
    plt.figure(figsize=(10, 6))

    plt.bar(dict1.keys(), dict1.values(), alpha=0.9, label='–¢—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–π MAE', color='red')

    plt.bar(dict2.keys(), dict2.values(), alpha=0.1, label='–¢–µ—Å—Ç–æ–≤—ã–π MAE', color='green')

    plt.xlabel('–ú–æ–¥–µ–ª–∏')
    plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ MAE')
    plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è MAE –º–æ–¥–µ–ª–µ–π –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –∏ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()

plot_dicts(results_train, results_test)
```

## –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –Ω–æ–≤—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö

```python
feature_1 = np.sin(X['feature_1_normal']) + np.log10(X['feature_2_binomial_1']) - \
    np.cos(X['feature_1_normal']) * np.exp(X['feature_3_binomial_2']) * np.sqrt(np.abs(X['feature_4_gamma']))

feature_2 = np.exp(X['feature_3_binomial_2']) * np.sqrt(abs(X['feature_2_binomial_1'])) * np.log(X['feature_1_normal']) / np.sqrt(abs(X['feature_4_gamma']))

feature_1_df = pd.DataFrame(data=feature_1, columns=['new_feature_1'])
feature_2_df = pd.DataFrame(data=feature_2, columns=['new_feature_2'])
```

```python
new_merged_df = merged_df.merge(feature_1_df, on='id')
new_merged_df = new_merged_df.merge(feature_2_df, on='id')

new_merged_df.head()
```

## –ü—Ä–æ–≤–µ—Ä–∫–∞ –º—É–ª—å—Ç–∏–∫–æ–ª–ª–∏–Ω–µ–∞—Ä–Ω–æ—Å—Ç–∏ –ø—Ä–∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–∏ –Ω–æ–≤—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö

```python
X_new = new_merged_df.drop(['feature_1_normal', 'score'], axis=1)
y_new = new_merged_df['score']
```

```python
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_new, y_new,
                                                            test_size=0.2,
                                                            random_state=0)
```

## üß™ –≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç—ã —Å –º–æ–¥–µ–ª—è–º–∏

–ü–æ–ø—Ä–æ–±—É–µ–º —Å–¥–µ–ª–∞—Ç—å —Å—Ç—ç–∫–∏–Ω–≥ –∏–∑ —Ç—Ä—ë—Ö —ç—Ç–∏—Ö –º–æ–¥–µ–ª–µ–π —Å –ª—É—á—à–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –∏ –∑–∞–º–µ—Ä–∏—Ç—å –∏—Ç–æ–≥–æ–≤—ã–π `MAE`

```python
from sklearn.ensemble import StackingRegressor

pipe_rf = Pipeline([('scaler', StandardScaler()),
                    ('RandomForestRegressor', RandomForestRegressor(bootstrap=True, max_depth=10,
                                                                    max_features='sqrt', min_samples_leaf=4,
                                                                    min_samples_split=2, n_estimators=200))])
pipe_catboost = Pipeline([('scaler', StandardScaler()),
                          ('CatBoostRegressor', CatBoostRegressor(depth=10, iterations=100,
                                                                  l2_leaf_reg=10, learning_rate=0.005,
                                                                  random_strength=0))])

estimators = [
    ('RF', pipe_rf),
    ('CBR', pipe_catboost)
]

final_estimator = DecisionTreeRegressor(criterion='absolute_error', max_depth=20,
                                        max_features='sqrt', min_samples_leaf=2,
                                        min_samples_split=5, min_weight_fraction_leaf=0.5)

stacking_model = StackingRegressor(
    estimators=estimators,
    final_estimator=final_estimator,
    cv=3
)
```

```python
%%capture
stacking_model.fit(X_train, y_train)
```

## –ò—Ç–æ–≥–æ–≤—ã–π MAE

```python
train_preds = stacking_model.predict(X_train)
test_preds = stacking_model.predict(X_test)

train_error = mean_absolute_error(y_train, train_preds)
test_error = mean_absolute_error(y_test, test_preds)

print(f"MAE –Ω–∞ —Ç—Ä–µ–π–Ω–µ –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ —Å—Ç–µ–∫–∏–Ω–≥–∞: {train_error}")
print(f"MAE –Ω–∞ —Ç–µ—Å—Ç–µ –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ —Å—Ç–µ–∫–∏–Ω–≥–∞: {test_error}")
```
